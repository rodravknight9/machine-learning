```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 100)
```

```python
# distance in km of the robber: d=2.5t
d_r = 2.5 * t 
```

```python
# distance in km of the police: d=3(t-5)
d_p = 3 * (t - 5)
```

```python
fig, ax = plt.subplots()
ax.plot(t, d_r, 'r-', label='Robber')
ax.plot(t, d_p, 'b-', label='Police')
ax.legend()
ax.set_xlim(0, 40)
ax.set_ylim(0, 100)
ax.set_xlabel('Time (h)')
ax.set_ylabel('Distance (km)')
ax.set_title('A Bank Robber Catch by the Police')
plt.show()
```
